```python
df.duplicated().sum()
```

## **Data Transformation**

```python
# Convert 'age' to integer
df['age'] = df['age'].astype(int)
df
```

```python
# Rename 'selfemp' to 'self_employed'
df.rename(columns={'selfemp': 'self_employed'}, inplace=True)
df
```

```python
# Create a new column 'balance_to_income_ratio'
df['balance_to_income_ratio'] = df['expenditure'] / df['income']
df
```

##  **Value Counts and Unique Values**



```python
# Value counts for 'card' column
print(df['card'].value_counts())

# Unique values in 'owner' column
print(df['owner'].unique())
```

## **Grouping and Aggregations**

```python
# Average expenditure by card ownership
avg_expenditure_by_card = df.groupby('card')['expenditure'].mean()
print(avg_expenditure_by_card)

# Sum of income by number of dependents
income_by_dependents = df.groupby('dependents')['income'].sum()
print(income_by_dependents)
```

## **Merging DataFrames**

```python
# Sample additional DataFrame
df_additional = pd.DataFrame({
    'rownames': [1, 2, 3],
    'credit_score': [700, 650, 720]
})
# Merge on 'rownames'
df_merged = pd.merge(df, df_additional, on='rownames', how='left')
df_merged
```

## **Pivot Tables and Crosstabulation**

```python
# Pivot table: average expenditure by card and owner status
pivot_table = df.pivot_table(values='expenditure', index='card', columns='owner', aggfunc='mean')
print(pivot_table)

# Crosstab: frequency of card ownership by self-employment status
crosstab = pd.crosstab(df['card'], df['self_employed'])
print(crosstab)
```

## **Applying Functions**

```python
# Define a function to categorize income levels
def categorize_income(income):
    if income < 2.5:
        return 'Low'
    elif 2.5 <= income < 5:
        return 'Medium'
    else:
        return 'High'

# Apply the function to create a new column
df['income_category'] = df['income'].apply(categorize_income)
df
```

## **Multivariate Analysis**

```python
# Selecting three numerical columns from the dataset
numerical_columns = df.select_dtypes(include=['number']).columns
selected_columns = numerical_columns[:3]  # Picking the first three numerical columns

# Set the title
plt.figure(figsize=(10, 6))
plt.suptitle("21BCE0420 - Pair Plot with Regression")

# Plot the pair plot with regression
sns.pairplot(df, vars=selected_columns, kind="reg")
plt.show()
```

```python
# Selecting three numerical columns for the pair plot
selected_columns = ['income', 'expenditure', 'share']

# Set the title
plt.figure(figsize=(10, 6))
plt.suptitle("21BCE0420 - Pair Plot of Income, Expenditure, and Share by Ownership", fontsize=14, fontweight='bold')

# Setting style and creating pair plot
sns.set(style="ticks", color_codes=True)
sns.pairplot(df, height=2, vars=selected_columns, hue="owner")
plt.show()
```

```python
import pandas as pd
from scipy import stats

# Selecting two numerical columns for correlation
col1 = "income"
col2 = "expenditure"

# Compute Pearson correlation coefficient
corr = stats.pearsonr(df[col1], df[col2])

# Display results
corr_results = {
    "p-value": corr[1],
    "correlation": corr[0]
}
corr_results
```

```python
correlation = df.corr(method='pearson', numeric_only=True)
correlation
```

```python
#Visualize this correlation analysis using a heatmap.
#A heatmap is the best technique to make this look beautiful and easier to interpret:
sns.heatmap(correlation,xticklabels=correlation.columns,
            yticklabels=correlation.columns)
plt.title("21BCE0420 - CORRELATION HEATMAP", fontsize=14, fontweight='bold')
```
